```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#displays matplot plots in notebook itself
%matplotlib inline
```

```
height = [150, 160, 170, 190]
weight = [60, 70, 80, 90]

plt.plot(height, weight)
```

```
plt.plot(height, weight)
#adding title
plt.title('Relationship between height and weight')

#adding xlabel and ylabel
plt.xlabel('Height(cms)')
plt.ylabel('Weight(kgs)')
```

```
calories_burnt = [55,65,95,85]
plt.plot(calories_burnt)
plt.plot(weight)
```

```
#adding labels
plt.plot(calories_burnt)
plt.plot(weight)

plt.legend(labels=['Calories burnt', 'Weight'], loc='lower right')
```

```
plt.plot(calories_burnt)
plt.plot(weight)

plt.legend(labels=['Calories burnt', 'Weight'], loc='lower right')

#set labels for each person
plt.xticks(ticks=[0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4'])
```

```
#expnad the plot
plt.figure(figsize=(15,5))

plt.plot(calories_burnt)
plt.plot(weight)

plt.legend(labels=['Calories burnt', 'Weight'], loc='lower right')

#set labels for each person
plt.xticks(ticks=[0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4'])
```

```
#to be plotted as green bubbles
plt.plot(calories_burnt, 'go')
plt.plot(weight, 'y--')

plt.legend(labels=['Calories burnt', 'Weight'], loc='lower right')

#set labels for each person
plt.xticks(ticks=[0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4'])
```

We can use subplots() to fit more than one plot in a figure


```
#create 2x2 plots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))

ax[0,0].plot(calories_burnt, 'go')
ax[0,1].plot(weight, 'y--')

ax[0,0].set_title('Calories burnt')
ax[0,1].set_title('Weight')

ax[0,0].set_xticks(ticks=[0,1,2,3])
ax[0,1].set_xticks(ticks=[0,1,2,3])

ax[0,0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4'])
ax[0,1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4'])
```

```
# here we encountered that there are too many readings on y axis
# the plots can share y axis for better readability
```

```
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6), sharey=True)

ax[0,0].plot(calories_burnt)
ax[0,1].plot(weight, 'y--')

ax[0,0].set_title('Calories burnt')
ax[0,1].set_title('Weight')

ax[0,0].set_xticks(ticks=[0,1,2,3])
ax[0,1].set_xticks(ticks=[0,1,2,3])

ax[0,0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4'])
ax[0,1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4'])
```
